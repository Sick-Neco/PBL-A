```python
# import
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.dates as mdate
import seaborn as sns

pd.set_option('display.max_columns', None) # pandas

# dfOur World in Data
df = pd.read_json("owid-covid-data.json")

df
```

```python
df.sort_values(by = 'max_correl',ascending=False,axis=1)
```

```python
cancer = pd.read_csv("lung-cancer-deaths-per-100000-by-sex-1950-2002.csv")

cancer
```

```python
cancer_insert = cancer[cancer['Year'] == 2016]
```

```python
cancer_insert.set_index("Code",inplace=True)

cancer_insert
```

```python
cancer_insert = cancer_insert.T
```

```python
cancer_insert
```

```python
df01 = pd.concat([df, cancer_insert])
df01 = df01.drop(['Entity','Year'])
```

```python
df01 
```

```python
countries = df.columns # 
date_keys = pd.DataFrame(df['USA']['data']).columns # datacolumn
date_keys
```

```python
df2 = pd.DataFrame(columns = countries, index = date_keys) # indexcolumndf2

# 
for country in countries:
    df_country = pd.DataFrame(df[country]['data'])
    latest_data = {}

    # 
    for column in df_country.columns:
        last_valid_index = df_country[column].last_valid_index()
        if last_valid_index is not None:
            latest_data[column] = df_country.loc[last_valid_index, column]
        else:
            latest_data[column] = np.nan

    # DataFrame
    df2[country] = pd.Series(latest_data)

# objectfloat
dft  = df01.T.apply(pd.to_numeric, errors='ignore')
dft2 = df2.T.apply(pd.to_numeric, errors='coerce')

# 
df_all = pd.concat([dft,dft2], axis=1)

df_all # df_all
```

```python
df_all = df_all.dropna(subset=['age-standardized_death_rate_per_100k_male'])
```

```python
df_all
```

```python
keys = ['continent', 'gdp_per_capita', 'human_development_index', 'female_smokers', 'male_smokers', 
        'age-standardized_death_rate_per_100k_male' , 'age-standardized_death_rate_per_100k_female',
        'total_cases_per_million' , 'total_deaths_per_million' , 'people_fully_vaccinated_per_hundred']
df_sb = df_all.filter(items=keys)
df_sb = df_sb.dropna(how = 'any')
df_sb
```

```python
sns.set(style='darkgrid')
sns.pairplot(df_sb, hue='continent')
plt.show()
```

```python
cor = df_sb.corr()
line_plot = sns.heatmap(cor, cmap= sns.color_palette('coolwarm', 10), annot=True,fmt='.2f', vmin = -1, vmax = 1)
figure = line_plot.get_figure()
figure.savefig('lung.png')
```
